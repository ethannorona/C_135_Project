```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("filtered_stars.csv")
df.head()
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
name = df['Name'].tolist()
distance = df['Distance'].tolist()
mass = df['Mass'].tolist()
radius = df['Radius'].tolist()
gravity = df['Gravity'].tolist()
```

```python
plt.figure(figsize = (10,5))
plt.title("Stars Solar Mass")
plt.bar(name[0:9], mass[0:9])
```

```python
plt.figure(figsize = (10,5))
plt.title("Stars Solar Radius")
plt.bar(name[0:9], radius[0:9])
```

```python
plt.figure(figsize = (10,5))
plt.title("Stars Gravity(m/s^2)")
plt.bar(name[0:9], gravity[0:9])
```

```python
plt.figure(figsize = (10,5))
plt.title("Stars Solar Distance")
plt.bar(name[0:9], distance[0:9])
```
